```python
import tensorflow as tf
from sklearn import datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##### 对鸢尾花数据集进行训练

```python
x_train=datasets.load_iris().data
y_train=datasets.load_iris().target
```

```python
y_train
```

```python
# 对数据集进行乱序
np.random.seed(1)
np.random.shuffle(x_train)
np.random.seed(1)
np.random.shuffle(y_train)
tf.random.set_seed(1)
```

```python
x_train
```

```python
y_train
```

```python
#  搭建网络结构
#  3 代表神经元的个数，activation表示使用什么样的激活函数，kernel_regularizer表示使用什么正则化方法
model=tf.keras.models.Sequential([
    tf.keras.layers.Dense(3,activation='softmax',kernel_regularizer=tf.keras.regularizers.l2())
])
```

```python
# 配置训练方法
#  optimize 选择SGD优化器,并且lr设置学习率为0.1
#  loss选择SparseCategoricalCrossentropy作为损失函数 --- 多分类交叉熵损失函数，由于softmax输出是概率分布而不是原始输出，将from_logits设置为False
#  由于鸢尾花数据集的target是[0,1,2]而神经网络的传播是概率分布,所以选择metrics=['sparse_categorical_accuracy']作为评测标准
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.1),
             loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
             metrics=['sparse_categorical_accuracy'])
```

```python
# 在fit中执行训练的过程,batch_size表示训练时一次喂入神经网络多少组数据,epochs表示数据集迭代的次数,
# validation_split表示数据集中多少比例的数据作为测试集,validation_freq表示迭代多少次之后进行一次评测
model.fit(x_train, y_train, batch_size=32, epochs=500, validation_split=0.2, validation_freq=20)
 
model.summary()
```

#### 使用函数封装模块实现鸢尾花数据集的神经网络训练

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras import Model
```

```python
x_train = datasets.load_iris().data
y_train = datasets.load_iris().target
 
np.random.seed(2)
np.random.shuffle(x_train)
np.random.seed(2)
np.random.shuffle(y_train)
tf.random.set_seed(2)
```

```python
class IrisModel(Model):
    def __init__(self):
        super(IrisModel,self).__init__()
        self.d1=Dense(3,activation='softmax',kernel_regularizer=tf.keras.regularizers.l2())
    def call(self,x):
        y=self.d1(x)
        return y
```

```python
model = IrisModel()
 
model.compile(optimizer=tf.keras.optimizers.SGD(lr=0.1),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
              metrics=['sparse_categorical_accuracy'])
 
model.fit(x_train, y_train, batch_size=32, epochs=500, validation_split=0.2, validation_freq=20)
model.summary()
```
